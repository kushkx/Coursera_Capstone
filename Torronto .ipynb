# Scraping Data from Wikipedia

## Assign url

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M' 
url
```

## Import pandas

```python
import pandas as pd
```

## Read HTML webpage into pandas

```python
df = pd.read_html(url, header = 0)
df
```

## Check Number of tables on the webpage

```python
len(df)
```

## Select first table on the web page and assign it to torronto

```python
  df[0].head()
```

```python
torronto = df[0]
```

## Data Cleaning

```python
torronto.dropna(subset =["Neighborhood"],axis = 0,inplace = True)
```

```python
torronto.reset_index(inplace = True)
```

```python
torronto.drop(['index'],axis = 1,inplace = True)
```
